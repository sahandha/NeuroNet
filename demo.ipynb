```python
from NeuroNet import *
%config InlineBackend.figure_formats=['svg']
plt.rcParams["figure.figsize"] = (13, 2)
```

```python
tend = 400
dt   = 0.1
```

```python
n1 = Neuron(1,dt=dt,tend=tend,a=0.8,b=0.7,tau=12.5,I=0.5)
n2 = Neuron(2,dt=dt,tend=tend,a=0.6,b=0.9,tau=11  ,I=0.7)
```

```python
b = Brain(neurons=[n1,n2],dt=dt,tend=tend,)
b.Simulate()
```

```python
b._Neurons[0].PlotState(states={0:"V",1:"w"},legend=["Voltage","Current"])
b._Neurons[1].PlotState(states={0:"V",1:"w"},legend=["Voltage","Current"])
```

```python
plt.rcParams["figure.figsize"] = (13, 4)
plt.suptitle('Input to cells')
plt.subplot(2,1,1)
plt.plot(n1._Time, n1._II)
plt.ylabel("n1 input")
plt.subplot(2,1,2)
plt.plot(n2._Time, n2._II)
plt.xlabel("Time")
plt.ylabel("n2 input")
plt.show()
```
