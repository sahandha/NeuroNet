```python
from NeuroNet import *
%config InlineBackend.figure_formats=['svg']
```

```python
n1 = Neuron(1,a=0.8,b=0.7,tau=12.5,I=0.5)
n2 = Neuron(2,a=0.6,b=0.9,tau=11,I=0.7)
```

```python
b = Brain(neurons=[n1,n2])
b.Simulate()
```

```python
plt.rcParams["figure.figsize"] = (16, 2)
b._Neurons[0].PlotState(states={0:"V",1:"w"},legend=["Voltage","Current"])
b._Neurons[1].PlotState(states={0:"V",1:"w"},legend=["Voltage","Current"])
```
