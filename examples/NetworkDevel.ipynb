```python
from NeuroNet import *
%config InlineBackend.figure_formats=['svg']
plt.rcParams["figure.figsize"] = (13, 2)
import time
```

```python
tend = 400
dt   = 0.01

numNeurons = 100
displaynum = min(10,numNeurons)
neurons = []


for i in range(numNeurons):
    if i==0:
        I=1
    else:
        I=0
    n = Neuron(i,dt=dt,tend=tend,a=random.uniform(0.6,0.9),b=random.uniform(0.5,0.9),tau=random.uniform(10,14),I=I)
    n.Initialize([-2,0])
    n._NoiseVariance = 0.00
    n.SetFlow(n.FHNFlow)
    n.SetSynapseLimit(random.uniform(980,1020))
    neurons.append(n)
    
b = Brain(neurons=neurons,dt=dt,tend=tend,connectionscale=100000)
for i in tnrange(4000,desc='Network Devel'):
    b.DevelopSynapseNetwork()
```

```python
n0 = b._Neurons[0]
n1 = b._Neurons[1]
n2 = b._Neurons[2]
n2._SynapseCount
n2._SynapticStrength
```

```python
plt.rcParams["figure.figsize"] = (8,8)
vis = Visualization(network=b._Network,neurons=b._Neurons, synapsecount=b._SynapseCountHistory)
#vis.DrawNetwork(edgelabels=False)
```
